#  Student Pass/Fail Prediction Using Decision Tree

This notebook demonstrates how to build and visualize a Decision Tree model using `scikit-learn` to classify student outcomes (Pass or Fail) based on Science scores.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
# Student dataset
data = {
    'Science_Score': [35, 45, 50, 55, 60, 70, 85],
    'Result': ['Fail', 'Fail', 'Pass', 'Pass', 'Pass', 'Pass', 'Pass']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display dataset info
print("STUDENT DATASET OVERVIEW\n")
print("Number of samples (students):", len(df))
print("\nData Description:\n", df.describe())
print("\nData Preview:\n", df)
```

```python
# Convert target column to numeric values
df['Result'] = df['Result'].map({'Fail': 0, 'Pass': 1})

# Split features and target
X = df[['Science_Score']]
y = df['Result']

# Train decision tree classifier
model = DecisionTreeClassifier(criterion='entropy', random_state=0)
model.fit(X, y)
```

```python
# Plot the decision tree
plt.figure(figsize=(8, 6))
plot_tree(model,
          feature_names=['Science_Score'],
          class_names=['Fail', 'Pass'],
          filled=True,
          rounded=True)
plt.title("Decision Tree - Student Pass/Fail Prediction")
plt.tight_layout()
plt.show()
```
